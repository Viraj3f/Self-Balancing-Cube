```python
%matplotlib notebook
import numpy as np
import matplotlib.pyplot as plt
from cube_system import CubeSystem
```

```python
cs = CubeSystem()
cs.current_theta_b = np.deg2rad(5)
time, theta_b, theta_w, phi_b, phi_w, I = cs.simulate(
    max_time=10, 
    h=0.001,
    position_gains=(-80, -230, -8))

plt.figure()
plt.plot(time, np.degrees(theta_b), "-")
plt.title("Body Angle in [deg]")
```

```python
plt.figure()
plt.plot(time[:-1], I, "-")
```
